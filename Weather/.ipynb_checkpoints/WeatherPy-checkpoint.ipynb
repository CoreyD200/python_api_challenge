# WeatherPy

Intent: Evaluate various weather data points across the world. Attempt to identify correlation between tempertaure, humidity, cloudiness and windspeed across various longitudes and latitudes

```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

from config import APIKEY

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Find 500 random cities across the globe 

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # Replace spaces with %20 to create url correctly 
    #city = city.replace(" ", "%20")
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:

        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## API calls to OpenWeatherMap

```python
# use weather data from OpenWeatherMap API

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + APIKEY + "&q="

```

```python
# get a json dump to figure out which keys to use to get the values
response = requests.get(url+"Denver")
data=response.json()
print(json.dumps(data, indent=4, sort_keys=True))
```

```python
#Print values to confirm
print(data["name"])
print(data["main"]["humidity"])
print(data['coord']['lat'])
print(data["main"]["temp"])
print(data["clouds"]["all"])
print(data["wind"]["speed"])
```

```python
# Create empty lists to append the API data into lists 
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# Set start for counter
record = 1
failed = 0

# Log file print statement
print(f"Beginning Data Retrieval")
print(f"-------------------------------")

#Loop through the cities in the city list 
for city in cities:  
    
    # Try statement to append calls where value is found 
    # Not all calls return data as OpenWeatherMap will not have have records in all the cities generated by CityPy module
    try: 
        response = requests.get(url + city).json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_number =(response['id'])
        
        print(f"Processing Record {record} | city#{city_number}:{city} ")
                
        # Increase counter by one 
        record= record + 1
        
        
    # If city not found output the city name and that it is being skipped, also account add 1 to the failed list
    except:
        failed = failed + 1
        print(f'{city} not found. Moving on to next record')
    #continue

 
record = record -1    
print('-------------------------')
print(f'Data Retrieval Completed')
print(f"Successfully retrieved data for {record} cities")
print(f'Unable to retrieve data for {failed} cities from openweathermap.org')
print('-------------------------')
```

```python
#create dictionary of values to add to dataFrame
data = {"City": city_name,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Lat":lat, 
    "Lng":lng, 
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
    }

#build DataFrame
weatherDF= pd.DataFrame(data)
weatherDF.head()
```

```python
# save dataframe to a csv file 
weatherDF.to_csv('WeatherData.csv', index=False)
```

## Northern and Southern hemisphere viewed individually

```python
#create seperate DF's for each hemisphere
northDF=weatherDF.loc[weatherDF['Lat']>=0]
southDF=weatherDF.loc[weatherDF['Lat']<0]
```

```python
def lin_regress(x_val,y_val):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_val, y_val)
    regress_values = x_val * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
```

```python
print(lin_regress)
```

### Temps


```python
x_val = northDF['Lat']
y_val = northDF['Max Temp']
#lin_regress
slope, intercept, rvalue, pvalue, stderr = linregress(x_val, y_val)
regress_values = x_val * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

cm = plt.cm.get_cmap('RdYlBu_r')

plt.scatter(x_val,y_val, c=y_val,vmin=-45, vmax =105, s=35, cmap=cm, edgecolors='grey')
plt.plot(x_val,regress_values,"black")
plt.annotate(line_eq,(0,-20),fontsize=15,color="Black")
plt.title('Temps vs Latitute (N)')
plt.xlabel('Northern Latitude')
plt.ylabel('Max Temp in °F')
plt.colorbar()
print(f"The r-squared is: {rvalue}")
plt.grid(axis = 'x')

#save chart image to images folder
plt.savefig("Images/fig5.png")


plt.show()
```

Strong correlation between temperatures and latitude in the Northern Hemisphere.
